```python
# Initialize an empty list to store student names
students = []

# Get 5 student names from the user
for i in range(5):
    name = input(f"Enter the name of student {i+1}: ")
    students.append(name)

# Display all student names in uppercase
print("\nStudent names in uppercase:")
for name in students:
    print(name.upper())
```

```python
# Initialize an empty list to store words
words = []

# Get 5 words from the user
for i in range(5):
    word = input(f"Enter word {i+1}: ")
    words.append(word)

# Display words that start with 'b' or 'B'
print("\nWords that start with 'b' or 'B':")
for word in words:
    if word.lower().startswith('b'):
        print(word)
```

```python
# Initialize an empty list to store the numbers
numbers = []

# Get 10 numbers from the user
for i in range(10):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

# Display numbers greater than 10 and less than 20
print("\nNumbers greater than 10 and less than 20:")
for num in numbers:
    if 10 < num < 20:
        print(num)
```

```python
colors = ["pink", "yellow", "green", "gold", "white", "gray", "green", "brown", "blue"]

newcolor = []

for i in colors:
    if "e" in i:
        newcolor.append(i)

print(newcolor)
```

```python
colors = ["pink", "yellow", "green", "gold", "white", "gray", "green", "brown", "blue"]

newcolor = [color for index, color in enumerate(colors) if "e" in color]

print(newcolor)
```

```python
# List containing numbers from 1 to 20
numbers = list(range(1, 21))

# Using list comprehension to filter even numbers
even_numbers = [num for num in numbers if num % 2 == 0]

# Printing the result
print("Even numbers:", even_numbers)
```
