```python
from datetime import date

d0 = date(2000, 3, 29)
d1 = date(2024, 4, 1)

years = d1.year - d0.year
months = d1.month - d0.month
days = d1.day - d0.day

if days < 0:
    months -= 1
    days += (d1.replace(year=d1.year, month=d1.month-1) - d1.replace(year=d1.year, month=d1.month-2)).days

if months < 8:
    years -= 1
    months += 12

print(f"Years: {years}, Months: {months}, Days: {days}")
```

```python
from datetime import datetime

def calculate_age(birth_date):
    today = datetime.today()
    age_in_years = today.year - birth_date.year
    age_in_months = today.month - birth_date.month
    age_in_days = today.day - birth_date.day

    if age_in_days < 0:
        age_in_months -= 1


    if age_in_months < 0:
        age_in_years -= 1
        age_in_months += 12

    total_days = (today - birth_date).days
    total_hours = total_days * 24
    total_minutes = total_hours * 60
    total_seconds = total_minutes * 60

    return age_in_years, age_in_months, age_in_days, total_days, total_hours, total_minutes, total_seconds

birth_date_input = input("Enter your birthdate (YYYY-MM-DD): ").replace(" ", "")
birth_date = datetime.strptime(birth_date_input, "%Y-%m-%d")

years, months, days, total_days, hours, minutes, seconds = calculate_age(birth_date)

print(f"Age: {years} years, {months} months, {days} days")
print(f"Total days: {total_days}")
print(f"Total hours: {hours}")
print(f"Total minutes: {minutes}")
print(f"Total seconds: {seconds}")

```
